QUESTION NO 6

```python

quiz = {
    "easy": [("2+2?", "4")],
    "medium": [("len('Python')?", "6")],
    "hard": [("bool('False')?", "True")]
}

score_rule = {
    "easy": (1, -0.25),
    "medium": (2, -0.5),
    "hard": (3, -1)
}

name = input("Name: ")
score = 0
seed = len(name)

for level in quiz:
    q, ans = quiz[level][seed % len(quiz[level])]

    print("\n", level.upper(), ":", q)
    user = input("Answer: ")

    if user == ans:
        score += score_rule[level][0]
        print("Correct")
    else:
        score += score_rule[level][1]
        print("Wrong")

print("\nFinal Score:", score)
```

QUESTION NO 7

```python
expenses = []

BUDGET = 5000

while True:
    print("\n--- Personal Expense Tracker ---")
    print("1. Add Expense")
    print("2. Show Total Expense")
    print("3. Category-wise Expense")
    print("4. Highest Expense Day")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        date = input("Enter date (DD-MM): ")
        category = input("Enter category: ")
        amount = int(input("Enter amount: "))

        expenses.append([date, category, amount])
        print("Expense Added!")

    elif choice == "2":
        total = 0
        for e in expenses:
            total = total + e[2]

        print("Total Expense:", total)

        if total > BUDGET:
            print("âš  Budget Exceeded!")

    elif choice == "3":
        categories = {}

        for e in expenses:
            if e[1] in categories:
                categories[e[1]] = categories[e[1]] + e[2]
            else:
                categories[e[1]] = e[2]

        print("Category-wise Expense:")
        for c in categories:
            print(c, ":", categories[c])

    elif choice == "4":
        day_total = {}

        for e in expenses:
            if e[0] in day_total:
                day_total[e[0]] = day_total[e[0]] + e[2]
            else:
                day_total[e[0]] = e[2]

        max_day = ""
        max_amount = 0

        for d in day_total:
            if day_total[d] > max_amount:
                max_amount = day_total[d]
                max_day = d

        if max_day != "":
            print("Highest Expense Day:", max_day, "-", max_amount)
        else:
            print("No expenses found")

    elif choice == "5":
        print("Thank you!")
        break

    else:
        print("Invalid choice")
```
